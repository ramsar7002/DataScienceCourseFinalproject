```python
arr = list(df['floors'])

for i in range(len(arr)):
    if '+' in str(arr[i]):
        res = arr[i].split('+')
        if len(res)==2:
            arr[i]= res[0]
        else:
            for item in res:
                if 'קרקע' in item:
                    arr[i] = 0
                if 'שלישית' in item:
                    arr[i] = 0
        
        
for i in range(len(arr)):
    if '+' in str(arr[i]):
        arr[i]=-2
        
df['floors']=arr
    
        
```
