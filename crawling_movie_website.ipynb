```python
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
from fake_useragent import UserAgent


driver = webdriver.Chrome()

user_scores = []
website_scores = []
movie_links = []
movie_names = []
release_years = []
directors = []
genres = []
durations = []

links = []
for i in range(0,140):
    str=f'https://www.metacritic.com/browse/movies/score/metascore/all/filtered?page={i}'
    links.append(str)

for link in links:
    options = Options()
    ua = UserAgent()
    userAgent = ua.random
    options.add_argument(f'user-agent={userAgent}')
    driver = webdriver.Chrome(chrome_options=options)

    driver.get(link)
    els = driver.find_elements_by_css_selector('.title')
    for el in els:
        if el.get_attribute('href') is not None:
            movie_links.append(el.get_attribute('href'))
    i = 0
    for movie in movie_links:
        driver.get(movie)

        score = driver.find_elements_by_css_selector('.metascore_anchor')
        name = driver.find_element_by_css_selector('.product_page_title')
        year = driver.find_element_by_css_selector('.release_year')
        director = driver.find_element_by_css_selector('.director')
        genre = driver.find_element_by_css_selector('.genres')
        duration = driver.find_element_by_css_selector('.runtime')

        website_scores.append(score[0].text)
        user_scores.append(score[1].text)
        movie_names.append(name.text)
        release_years.append(year.text)
        directors.append(director.text)
        genres.append(genre.text)
        durations.append(duration.text)

        print(website_scores[i])
        print(user_scores[i])
        print(movie_names[i])
        print(release_years[i])
        print(directors[i])
        print(genres[i])
        print(durations[i])
        print()
        i += 1
```
