```python
df = pd.read_csv('houses2.csv')
```

```python
df = df[df.rooms!=0]
df.head()
```

```python
df.replace(' ', np.nan, inplace= True)
```

```python
df.dropna(subset=['addresses', 'prices'], inplace=True)
df.head()
```

```python
df.info()
```

```python
df.drop('changes', inplace= True, axis=1)
df.info()
```

```python
df =df[df.kinds!="חנות"]
df =df[df.kinds!="תעשיה"]
df =df[df.kinds!="מלאכה"]
df =df[df.kinds!="משרד"]
df =df[df.kinds!="מחסנים"]
df.info()
```
